```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,200)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def signum(x):
    return np.sign(x)

def bipolar_sigmoid(x):
    return (1 - np.exp(-x)) / (1 + np.exp(-x))

def binary_sigmoid(x, threshold=0.5):
    return 1 if x >= threshold else 0

def threshold_function(x, threshold=0):
    return 1 if x >= threshold else 0

def step_function(x):
    return 1 if x >= 0 else 0

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x >= 0, x, alpha * x)

def tanh(x):
    return np.tanh(x)

plt.figure(figsize=(15,10))
plt.subplot(3,3,1)
plt.title("Roll no:21051016 Sigmoid Activation")
plt.plot(x,sigmoid(x))

plt.subplot(3,3,2)
plt.title("Roll no:21051016 ReLU Activation")
plt.plot(x,relu(x))

plt.subplot(3,3,3)
plt.title("Roll no:21051016 Leaky ReLU Activation")
plt.plot(x,leaky_relu(x))

plt.subplot(3,3,4)
plt.title("Roll no:21051016 Tanh Activation")
plt.plot(x,tanh(x))
 
plt.subplot(3,3,5)
plt.title("Roll no:21051016 Bipolar sigmoid")
plt.plot(x,bipolar_sigmoid(x))

plt.subplot(3,3,6)
plt.title("Roll no: 21051016 Binary Sigmoid")
plt.plot(x,[binary_sigmoid(i) for i in x])

plt.subplot(3,3,7)
plt.title("Roll no:21051016 Threshold Function")
plt.plot(x,[threshold_function(i) for i in x])

plt.subplot(3,3,8)
plt.title("Roll no:21051016 Step Function")
plt.plot(x,[step_function(i) for i in x])

plt.tight_layout()
plt.show()
```
